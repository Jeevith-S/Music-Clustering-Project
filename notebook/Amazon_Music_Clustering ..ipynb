```python
features = [
    'danceability',
    'energy',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo',
    'duration_ms'
]
```

```python
X = one[features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

one['PC1'] = X_pca[:, 0]
one['PC2'] = X_pca[:, 1]

print(one[['PC1', 'PC2']].head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(one['PC1'], one['PC2'])

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Visualization")

plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sse = []

for i in range(1,10):

    model = KMeans(n_clusters=i)

    model.fit(X_scaled)

    sse.append(model.inertia_)

plt.plot(range(1,10), sse)

plt.xlabel("k")
plt.ylabel("sse")

plt.show()
```

```python
kmeans = KMeans(n_clusters=5)

one['cluster'] = kmeans.fit_predict(X_scaled)

print(one['cluster'].head())
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=one['PC1'],
    y=one['PC2'],
    hue=one['cluster']
)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, one['cluster'])

print(score)
```

```python
one.groupby('cluster')[features].mean()
```
